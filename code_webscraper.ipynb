```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.leonardsguide.com/us-airport-codes.shtml')

#Parsing Request HTML Content
soup = BeautifulSoup(r.content, 'html.parser')

s = soup.find('div')
airport_codes = soup.find_all('td')

type(airport_codes)
```
